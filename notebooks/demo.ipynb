```python
import sys  
import os

# include workspace in path for uuv_mission imports
workspace_dir = os.path.abspath('..')
sys.path.append(workspace_dir)

# Import relevant modules
from uuv_mission.control import PD_Controller
from uuv_mission.dynamic import Submarine,Mission,ClosedLoop
```

# Simulate closed-loop from csv file
Simulate the controller dynamics using the mission.csv mission

```python
sub = Submarine()
controller = PD_Controller(Kp=0.08,Kd=0.85)
closed_loop = ClosedLoop(sub, controller)
mission = Mission.from_csv("..\data\mission.csv")

trajectory = closed_loop.simulate_with_random_disturbances(mission)
trajectory.plot_completed_mission(mission)
```

# Simulate closed-loop for random mission
Simulate the controller dynamics using a randomly generated mission

```python
# randomly generate a new mission
duration = 100
scale = 10
mission = Mission.random_mission(duration,scale)
```

```python
# simulate the new mission
sub = Submarine()
controller = PD_Controller(Kp=0.08,Kd=0.85)
closed_loop = ClosedLoop(sub, controller)

variance = 0.5 # variance of random disturbances (0.5 by default)

trajectory = closed_loop.simulate_with_random_disturbances(mission,variance)
trajectory.plot_completed_mission(mission)
```
